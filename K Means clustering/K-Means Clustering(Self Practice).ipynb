```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
#the data is a tuple (a,b) where 'a' is a 2D array of dimension (200,2) and 'b' is a 1-D array containing 200 elements 
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='viridis')
# plot shows that it consists of 4 blobs
```

```python
from sklearn.cluster import KMeans
```

```python
# defining the no. of clusters we want
clf = KMeans(n_clusters=4)
clf.fit(data[0])
```

```python
clf.cluster_centers_
```

```python
# original labels...we know this cuz we created the blobs ourselves else in unsupervised we dont know the original label.
data[1]
```

```python
# predicted labels
clf.labels_
```

```python
# making subplots to visualise the original and the predicted labels.
fig,(ax1,ax2) = plt.subplots(1,2, sharey =True,figsize=(10,6))

ax1.set_title('Predicted by K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=clf.labels_,cmap = 'rainbow')

ax2.set_title('Original labels')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
